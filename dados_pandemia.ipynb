```python
#IMPORTAÇÃO DAS BIBLIOTECAS
import pandas as pd
import numpy as np
```

```python
#IMPORTAÇÃO DOS DADOS DE VACINAÇÃO POR DATA EM CADA ESTADO
vacina = pd.read_csv(r"https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-states.csv")

vacina.head()
```

```python
#VERIFICAR OS DADOS DO DATASET
vacina.info()
```

```python
#TRANSFORMAR DATA DE object PARA DATETIME
vacina["date"] = pd.to_datetime(vacina["date"])
```

```python
#PREENCHER VALORES VAZIOS NA COLUNA DE VACINADOS COM NÚMERO 0
vacina["vaccinated"] = vacina["vaccinated"].fillna(0)
```

```python
#PREENCHER VALORES VAZIOS NA COLUNA DE VACINADOS SEGUNDA DOSE COM NÚMERO 0
vacina["vaccinated_second"] = vacina["vaccinated_second"].fillna(0)
```

```python
#PASSAR OS DADOS DE VACINADOS DE FLOAT PARA INT
vacina["vaccinated"] = vacina["vaccinated"].values.astype(np.int64)
```

```python
#PASSAR OS DADOS DE VACINADOS 2 DOSE DE FLOAT PARA INT
vacina["vaccinated_second"] = vacina["vaccinated_second"].values.astype(np.int64)
```

```python
#RETIRAR LINHAS DE TOTAL
vacina.drop(vacina[vacina["state"]=="TOTAL"].index,inplace=True)
```

```python
#FILTRAR APENAS AS COLUNAS IMPORTANTES
estados = vacina[["date","state","vaccinated","vaccinated_second"]]
```

```python
#PASSAR AS COLUNAS PARA PT-BR
estados.columns = ["DATA","ESTADOS","1° DOSE","2° DOSE"]
```

```python
#CONFIRMAR AS INFORMAÇÕES
estados.info()
```

```python
#PLOTAR UM PRÉ-VISUALIZAÇÃO DOS DADOS
estados
```

```python
#GERAR EXCEL NA PASTA ONDE ENCONTRA-SE O ARQUIVO PYTHON
estados.to_excel("estados.xlsx", index=False)
```
